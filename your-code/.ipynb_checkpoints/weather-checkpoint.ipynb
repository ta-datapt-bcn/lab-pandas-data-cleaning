## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import datetime as dt
```

```python
weather = pd.read_csv('weather-raw.csv')
```

```python
w = weather.melt(id_vars=['id','year','month','element'], var_name='day', value_name='temp')
```

```python
w.pivot_table(index=['id','year','month','day'], values='temp', columns='element').reset_index()
```

```python
w.pivot_table(index=['id','year','month','day'], values='temp', columns='element').reset_index().assign(day=w.day.apply(lambda x: x.replace('d','')).astype('int'))
```

```python
weather = w.pivot_table(index=['id','year','month','day'], values='temp', columns='element').reset_index().assign(day=w.day.apply(lambda x: x.replace('d','')).astype('int'))
```

```python
weather['date'] = weather.apply(lambda row: dt.date(row['year'],row['month'],row['day']),axis=1)
weather['date'].dtype
```

```python
weather['date'] = pd.to_datetime(weather.apply(lambda row: dt.date(row['year'],row['month'],row['day']),axis=1))
```

```python
weather.info()
```

```python
weather['tmin'].plot()
weather['tmax'].plot()
```

```python
weather['month_name']=weather['date'].dt.strftime('%B')
weather
```

```python
weather.groupby(['month_name']).mean()
```
